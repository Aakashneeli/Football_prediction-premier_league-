```python
#Cleaning our data for machine learning
#as we know machine learning models can only work with numerical data, so only float64 and in64 , cant with object type data
df.dtypes
```

```python
#first we'll convert date column into the date time data type in pandas because right now its stored as a string.

df['date'] = pd.to_datetime(df['date']) # here its not actually creating a new column, its converting the existing column into a datetime dtype and then overwriting the existing column with the datetime
df.dtypes 
# dtype of date changed to datetime
```

```python
#Creating Features

#first we'll create a venue code column , for home and away as they are categorical,
#and in football venue is very important , you may have heard of home advantage.
df['venue_code'] = df['venue'].astype('category').cat.codes # 0-Away, 1-Home

#now we'll create unique codes for each opponent. same process
df['opp_code'] = df['opponent'].astype('category').cat.codes

#maybe some teams play better at certain times of day
df['hour'] = df['time'].str.replace(':.+', '', regex = True).astype('int')
#day code 
df['day_code'] = df['date'].dt.dayofweek
df['target'] = (df['result'] == 'W').astype('int') # what (df['result'] == 'W') this'll do is it gives us a boolean True for W and False for L and D , and we use astype to convert it into 0 and 1
df
```

```python
#creating our initial machine learning model
from sklearn.ensemble import RandomForestClassifier
# A random forest is series of decision trees and each tree has different parameters
rf = RandomForestClassifier(n_estimators = 50, min_samples_split = 10, random_state = 1)# n_estimators is the number of individual decison trees we want to train, the more the number is the longer itll take but itll be more acuurate 

# spliting training and test data 
#this data is time series data so to make sure all of the data in the test set comes after the train set 
train = df[df['date'] < '2022-01-01']
test = df[df['date'] > '2022-01-01']

predictors = ['venue_code', 'opp_code', 'day_code', 'hour']
y = train['target']

#fitting the model 
rf.fit(train[predictors], y)
preds = rf.predict(test[predictors])
```

```python
#check the accuracy 
from sklearn.metrics import accuracy_score
acc = accuracy_score(test['target'], preds) # accuracy score tells us how overall correctness of the model, both postive and negative. no.of crct preds/total no.crct preds
acc
```

```python
# lets see in which situations our accuracy was high and low 
combined = pd.DataFrame(dict(actual = test['target'], predictions = preds))

# we can create a crossatab that'll show us, when we predicted a 0 or 1 what actually happened
pd.crosstab(index = combined['actual'], columns = combined['predictions'])
```

```python
from sklearn.metrics import precision_score
precision_score(test['target'], preds) # precision score tells us the accuracy of the positive predictions made by the model. True +ve/ true+ve + false+ve
```

```python
#splitting our matches df by team.to calculate the rolling averages
grouped_matches = df.groupby('team')
group = grouped_matches.get_group('Manchester City')

#creating a function for rolling avg # chatgpt'd
def rolling_averages(group, cols, new_cols):
    group_sorted = group.sort_values('date')

    # Exclude the 'date' column from the rolling operation
    rolling_stats = group_sorted[cols].rolling(window=3, closed='left').mean()

    # Reset the index after rolling operation
    rolling_stats.reset_index(drop=True, inplace=True)

    # Assign the rolling averages to new columns in the DataFrame
    for col, new_col in zip(cols, new_cols):
        group[new_col] = rolling_stats[col]

    # Drop rows with NaN values (resulting from the rolling window)
    group = group.dropna(subset=new_cols)

    return group


cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt']
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols , new_cols)
```

```python
df_rolling = df.groupby('team').apply(lambda x: rolling_averages(x, cols, new_cols))
df_rolling
```
